**Q1 - decimal to binary (2)**

```python
def binary(n):
    if(n==0):
        s="0";
    else:
        s = "";
    while(n>0):
        if(n%2 == 0):
            s = "0"+s;
            n=n//2
        elif(n%2 != 0):
            s = "1"+s;
            n=n//2
    
    return s;
```

```python
print(binary(25))
```

**Q1 - decimal to octal (8)**

```python
def octal(n):
    if(n==0):
        s="0";
    else:
        s = "";
    while(n>0):
        s = str(n%8) +s
        n=n//8
    
    return s;
```

```python
print(octal(153))
```

**Q1 - decimal to hexadecimal (16)**

```python
def hex(n):
    if(n==0):
        s="0";
    else:
        s = "";
    while(n>0):
        if(n%16 <= 9):
            s = str(n%16) +s
        elif(n%16 > 9):
            if(n%16==10):
                s = "A" +s;
            elif(n%16==11):
                s = "B" +s; 
            elif(n%16==12):
                s = "C" +s; 
            elif(n%16==13):
                s = "D" +s;  
            elif(n%16==14):
                s = "E" +s; 
            elif(n%16==15):
                s = "F" +s; 
        n=n//16;
    return s;
```

```python
print(hex(1000))
```
